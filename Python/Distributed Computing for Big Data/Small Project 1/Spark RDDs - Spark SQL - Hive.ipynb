I transform each row of the *data_from_file* RDD from a string to a string array. The string is split into different elements with the *\t* delimiter. 

```python
import numpy as np
data_from_file_conv = data_from_file.map(lambda row: np.array(row.strip().split("\t")))
data_from_file_conv.take(5)
```

The first and third elements of the arrays that were created before are the *author name* and *title of paper* equivallently. Therefore, these 2 elements are seleced and conacatenated into a string, with a *\t* symbol in between and the string 'Author_Title:' in the beginning. 

```python
author_name = data_from_file_conv.map(lambda row: ("Author_Title:" + row[0] + "\t" + row[2]))
author_name.take(5)
```

I convert the delimiter (*\t*) separated string, named *author_name*, to an array of string elements, named *name_count*. <br>
I then create the *title_author* RDD object, which has the string arrays, but with a switched order of their elements. So, the title is the first element and the author is the second element.

```python
name_count = author_name.map(lambda row: (row.strip().split('Author_Title:')[1].split('\t', 1)))
title_author = name_count.map(lambda row: (row[1], row[0]))
title_author.take(5)
```

Join the *title_author* RDD with itsself on the title. This creates string arrays, whose first element is the title and the second element a pair of co-authors for that paper.

*(Note: After doing some exploration in the initial dataframe, I have seen that a title can be published in more than 1 journals, leading into having the combination of a title and an author in more than one row in some cases. In the instructions of this assignment it was mentioned that we consider two authors to be collaborators when they have a common title. Hence, in this assignment I have not taken into account the different journals that the same title could have been published. In other words, if author A and B published a paper with title T1 in journals J1 and J2, I will count their collaboration only once.)*

```python
author_pairs = title_author.join(title_author)
author_pairs.take(5)
```

Sort the value of the RDD, in other words sort the elements in the array, so that same author pairs appear in the same order 
and hence can be interpreted by Spark as the same value.

```python
author_pairs_sort= (author_pairs.map(lambda row: (row[0], sorted(row[1]))))
author_pairs_sort.take(5)
```

Remove the pairs, where authors formed a pair with themselves. For example, we are not interest in the pair ['Stavros Christodoulakis', 'Stavros Christodoulakis'].

```python
author_pairs_nosame =author_pairs_sort.filter(lambda row: row[1][0] != row[1][1])
author_pairs_nosame.take(5)
```

Convert the pairs to a tuple format and the select only the distinct rows from the RDD. The reason for selecting distinct rows is that pairs appeared in the RDD twice for each title after the join.

```python
author_pairs_dist = author_pairs_nosame.map(lambda row: (row[0], tuple(row[1]))).distinct()
author_pairs_dist.take(5)
```

Make the pairs the keys of the RDD, by bringing them in the beginning of the array, and assign the number 1 as their value.

```python
pairs_1 = author_pairs_dist.map(lambda row: (row[1], int(1)))
pairs_1.take(5)
```

Sort the RDD by keys (by pair of co-authors), to facilitate the reduce part afterwards.

```python
pairs_1_sort = pairs_1.sortByKey()
pairs_1_sort.take(5)
```

Group the RDD by the pair of co-authors and count the number of times that each pair appeared.

```python
pairs_reduced = pairs_1_sort.reduceByKey(lambda x, y: x + y)
pairs_reduced.take(5)
```

Sort the count by descending order and print the top 10 pairs of authors who published the largest number of papers together.

```python
pairs_reduced.sortBy((lambda x: x[1]), ascending = False).take(10)
```

# P2 - Spark SQL

The `author-large.txt` file is on my bucket from the previous exercise. I store its location to the `filename` object.

```python
filename = 'gs://sarant-st446-1/author-large.txt'
```

I load the data from `author-large.txt` file into a DataFrame called `author_large` and programmatically apply the schema into the data. I then read the data from the file and load them into a Dataframe.

```python
from pyspark.sql.types import *

#Create the schema to use for the author-large file
schema = StructType([
    StructField("author", StringType(), True),    
    StructField("journal", StringType(), True),
    StructField("title", StringType(), True),
    StructField("year", LongType(), True)
])

#Load the data from the file into a Dataframe
author_large = spark.read.csv(filename, header='false', schema=schema, sep='\t')
author_large.createOrReplaceTempView("author_large")
```

 #### Find the top 10 pairs of authors who published the largest number of papers together.
 
 The steps followed and included in the sql command are described here. I first join the **author_large** table with itself on the **title**. From the first table instance we get the first author and from the second table instance the second author of the pair. We have a `where` clause, where we make sure that we don't create a pair with the same author name. From this joint table we concatenate the two author names into 1 column by sorting their names in alphabetic order, using the functions `concat`, `least` and `greatest`. We name this column **pair_of_authors**. From the joined table we then select the distinct combination of the **title** and **pair_of_authors** columns. We needed to specify the `distinct`, because the result of the join would be 2 records for each pair of authors. We then `group` this table by the **pair_of_authors** and `count` the number of times that each pair appears in the table. We therefore create the **joint_publications** column on the new grouped table, which we also sort by `descending order` based on the number of joint publications that each pair of author had and select to show only the `top 10` pairs, without concatinating the results (`truncate =False`). 

```python
spark.sql("select pair_of_authors, count(*) as joint_publications \
                from(select distinct tab1.title, \
                       concat(least(tab1.author, tab2.author), ' , ', greatest(tab1.author, tab2.author)) as pair_of_authors\
                       from author_large tab1 join author_large tab2 \
                       on tab1.title =tab2.title\
                       where tab1.author!= tab2.author)\
                group by pair_of_authors order by count(*) desc").show(10, truncate =False)
```
